Matplotlib is a Python library used for data visualization. It provides a variety of customizable plots such as line plots, scatter plots, bar plots, histograms, and many others.

Matplotlib is widely used in data analysis and scientific research to visualize data in a meaningful way. It allows users to create complex plots with just a few lines of code, making it a valuable tool for data visualization.

Pyplot is a sublibrary of Matplotlib that provides a simple interface for creating plots. Here are five plots that can be plotted using the Pyplot module of Matplotlib:

Line Plot: A line plot is a graph that displays data as a series of points connected by lines. This type of plot is useful for displaying trends over time.

Scatter Plot: A scatter plot is a graph that displays data as a collection of points. This type of plot is useful for displaying relationships between variables.

Bar Plot: A bar plot is a graph that displays data as bars. This type of plot is useful for comparing values between different groups.

Histogram: A histogram is a graph that displays the distribution of a dataset. This type of plot is useful for understanding the shape of the data.

Pie Chart: A pie chart is a graph that displays data as slices of a pie. This type of plot is useful for showing the relative proportions of different categories.

# Quetion : 2

A scatter plot is a type of plot that displays the relationship between two variables. It is useful for identifying patterns and relationships in data. In a scatter plot, each data point is represented as a point on the graph, with the x-axis representing one variable and the y-axis representing the other variable.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y,color='red')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("scatter chart")
```

# Quetion : 3

The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to arrange multiple plots in a grid, where each plot represents a different subset of the data or a different visualization.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line Plot 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line Plot 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line Plot 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line Plot 4')

fig.suptitle('Four Line Plots')
plt.show()
```

# Quetion : 4

A bar plot is a type of plot that represents data as bars, where the length of each bar corresponds to the value of a particular category or group. It is commonly used to show comparisons between different categories or groups of data.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
plt.title("profit chart")
```
